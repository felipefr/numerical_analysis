```python
import numpy as np
import matplotlib.pyplot as plt

N = 20;

x = np.logspace(1,N,N)**-1;

y1 = ((x+1)**2-1)/x
y2 = x + 2

plt.plot(x,y1,'o')
plt.plot(x,y2,'r+')
plt.legend(['y1', 'y2'])
plt.xscale('log')
plt.yscale('log')
plt.grid()
```

```python
err12 = np.abs(y1-y2)

plt.plot(x,err12,'o')
plt.legend(['y1-y2'])
plt.grid()
plt.xscale('log')
plt.yscale('log')
```
