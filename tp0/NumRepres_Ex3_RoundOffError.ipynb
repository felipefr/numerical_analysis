```python
import numpy as np
from matplotlib import pyplot as plt

X=np.linspace(0.995,1.005,100)

Y1 = (X-1)**6;
Y2 = X**6 - 6*X**5 + 15*X**4 -20*X**3 + 15*X**2 - 6*X + 1

plt.plot(X,Y2,'r');
plt.plot(X,Y1,'b');
plt.legend(['Y1', 'Y2'])
```
